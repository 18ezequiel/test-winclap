# EDA

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/dataset.csv')
df.head(10)
```

```python
# Get general information about the DataFrame
print("\nGeneral information about the DataFrame:")
df.info()
```

```python
# Print out statistical summary
print("\nStatistical summary:")
df.describe()
```

```python
# Print out count of unique values
print("\nCount of unique values:")
df.nunique()
```

```python
# Grouping by 'campaign_name' and counting total values
total_values_by_campaign = df.groupby('campaign_name').count()

# Print out total count of values per campaign
print("\nTotal count of values per campaign:")
total_values_by_campaign
```

```python
# Select relevant columns to check for missing values
columns_to_check = ['impressions', 'clicks', 'installs', 'spend']

# Group by 'campaign_name' and calculate the count of missing values
missing_values_by_campaign = df.groupby('campaign_name')[columns_to_check].apply(lambda group: group.isna().sum())

# Print out count of missing values per campaign
print("\nCount of missing values per campaign:")
missing_values_by_campaign
```

```python
# Create a list to store the results
zero_counts = []

# Iterate through each campaign to count the values equal to 0 in specific columns
for campaign in df['campaign_name'].unique():
    campaign_df = df[df['campaign_name'] == campaign]
    zero_impressions = (campaign_df['impressions'] == 0).sum()
    zero_clicks = (campaign_df['clicks'] == 0).sum()
    zero_installs = (campaign_df['installs'] == 0).sum()
    zero_spend = (campaign_df['spend'] == 0).sum()
    zero_counts.append({'campaign_name': campaign, 'zero_impressions': zero_impressions,
                        'zero_clicks': zero_clicks, 'zero_installs': zero_installs, 'zero_spend': zero_spend})

# Create a DataFrame from the results list
zero_counts_df = pd.DataFrame(zero_counts)

# Display the DataFrame
zero_counts_df
```

```python
# Identify duplicated rows
duplicated_rows_all = df[df.duplicated(keep=False)]

# Print duplicated rows
print("Duplicated rows:")
print(duplicated_rows_all)
```

### Analysis of missing or null values vs. actual values

```python
import matplotlib.pyplot as plt

# Filter the DataFrame
columns_to_plot = ['summary_date', 'impressions', 'clicks', 'installs', 'spend']
df_filtered = df[columns_to_plot]

# Create bar plots for each column
for column in columns_to_plot:
    plt.figure(figsize=(10, 6))
    plt.title(f'Distinct Non-zero Values vs. Zero or Null Values in {column}')
    
    # Count distinct non-zero values and zero or null values
    values_not_zero = df_filtered[df_filtered[column] != 0][column].count()
    values_zero = df_filtered[df_filtered[column] == 0][column].count()
    
    labels = ['Distinct Non-zero Values', 'Zero or Null Values']
    
    values = [values_not_zero, values_zero]
    
    plt.bar(labels, values, color=['blue', 'red'])
    
    plt.show()
```

### Grouping Columns by Campaign

```python
# List of metrics
metrics = ['impressions', 'clicks', 'installs', 'spend']

# Iterate through each metric
for metric in metrics:

    # Group data by campaign and sum the metric
    metric_by_campaign = df.groupby('campaign_name')[metric].sum()

    plt.figure(figsize=(10, 6))
    metric_by_campaign.plot(kind='bar', color='blue')
    plt.xlabel('Campaign')
    plt.ylabel(metric.capitalize())  # Use the capitalized metric name
    plt.title(f'{metric.capitalize()} by Campaign')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

### Metrics

```python
# Group by campaign and calculate the required metrics
grouped_df = df.groupby('campaign_name').agg({
    'summary_date': 'first',   # Select the first date within the group
    'impressions': 'sum',       # Sum the impressions within the group
    'clicks': 'sum',            # Sum the clicks within the group
    'installs': 'sum',          # Sum the installs within the group
    'spend': 'sum'              # Sum the spend within the group
})
```

```python
# Group by campaign and calculate the required metrics
grouped_df = df.groupby('campaign_name').agg({
    'impressions': 'sum',
    'clicks': 'sum',
    'installs': 'sum',
    'spend': 'sum'
})

# Define a function to calculate CTR and CPI for each campaign
def calculate_metrics(row):
    # Calculate CTR: Clicks / Impressions (avoid division by zero)
    row['CTR'] = row['clicks'] / row['impressions'] if row['impressions'] != 0 else 0
    
    # Calculate CPI: Spend / Installs (avoid division by zero)
    row['CPI'] = row['spend'] / row['installs'] if row['installs'] != 0 else 0
    
    return row

# Apply the function to each row of the grouped DataFrame
grouped_df = grouped_df.apply(calculate_metrics, axis=1)

# Display the resulting DataFrame with calculated metrics
grouped_df
```
